```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("bird.csv")
df
```

### Line plot

```python
plt.plot(df["feml"][:])
plt.show()
```

### Scatter plot

```python
plt.scatter(df["huml"][:],df["humw"][:])
plt.xlabel('data of huml')
plt.ylabel('data of humw')
plt.title("Bird Dataset")
plt.show()
```

### Histogram

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df["type"] = le.fit_transform(df["type"])
df
```

```python
plt.hist(df["type"], width = 0.6)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

### Bar graph

```python
plt.bar(df["type"],df["tarw"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

### Pie chart

```python
plt.pie(df["huml"][:10],labels = df["humw"][:10])
plt.show()
```

### Box Plot

```python
plt.boxplot(df["type"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```
