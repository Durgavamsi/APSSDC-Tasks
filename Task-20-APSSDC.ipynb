```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/Glass_Classification.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.drop('Type',axis=1)
y=df['Type']
```

```python
x.head()
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.40, random_state = 45)
```

```python
x_train
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#applying forest model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc
```

```python
rfc.fit(x_train, y_train)
```

```python
Y_prediction = rfc.predict(x_test)
Y_prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_prediction, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_prediction, y_test)
```
