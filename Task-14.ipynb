### Task1
Replace null values "Age" coulumn with ffill()
Replace null values in "Cabin" and "Embarked" columns with most occure value

```python
import numpy as np
import pandas as pd
```

```python
titanic=pd.read_csv("titanic.csv")
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic['Age']=titanic['Age'].ffill()
```

```python
titanic
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Cabin'].value_counts()
```

```python
titanic['Cabin']=titanic['Cabin'].replace(np.nan,"G6")
```

```python
titanic
```

```python
titanic['Embarked']=titanic['Embarked'].replace(np.nan,"S")
```

```python
titanic
```

```python
titanic.isnull().sum()
```

### Task3
Display the plot of original data and robust scaler data

```python
data = pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv")
```

```python

data
```

```python
data.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
data
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs=RobustScaler()
```

```python
rs2=rs.fit_transform(data)
rs2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(9,5))
sns.kdeplot(data['TV'],ax=ax1)
sns.kdeplot(data['radio'],ax=ax1)
sns.kdeplot(data['newspaper'],ax=ax1)
sns.kdeplot(data['sales'],ax=ax1)

sns.kdeplot(rs2['TV'],ax=ax2)
sns.kdeplot(rs2['radio'],ax=ax2)
sns.kdeplot(rs2['newspaper'],ax=ax2)
sns.kdeplot(rs2['sales'],ax=ax2)
```
